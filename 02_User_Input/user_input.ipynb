```python
# calculate simple interest

p=int(input("Enter Principal AMount :: "))
r=float(input("enter Rate of Interest :: "))
n=int(input("Enter Years :: "))

si=(p*r*n)/100

print("Simple Interest is :: ",si)
```

```python
# min and max number 
a=int(input("Enter First Number :: "))
b=int(input("Enter Second Number :: "))
c=int(input("Enter Third Number :: "))
d=int(input("Enter Fourth Number :: "))

print("Maximum Number is :: ",max(a,b,c,d))
print("Minimum Number is :: ",min(a,b,c,d))
```

```python
# swap the number
a=int(input("Enter First Number :: "))
b=int(input("Enter Second Number :: "))

print("Before Swapping values of a and b are :: ",a,b)

a,b=b,a

print("After Swapping the Values of a and b are :: ",a,b)
```

```python
# calculate the power of any number 
base=int(input("Enter Base Number ::  "))
exp=int(input("Enter Exponential Number :: "))

res=base**exp
print("Result is :: ",res)
```

```python
# Count the total length of sentence and also the number of words in it

sen=input("Enter Sentence/String :: ")
# length of string
print("Length of String is :: ",len(sen))

# word in string 
print("Total Words in String is :: ",len(sen.split()))

# replacing 
print("Replaced String is :: ",sen.replace(" ","="))
print("Replaced String is :: ",sen.replace("India","Hindustan"))
```
