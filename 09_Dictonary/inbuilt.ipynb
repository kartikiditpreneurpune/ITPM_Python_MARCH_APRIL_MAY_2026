```python
# Shallow Copy :
# Problem :::
student={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}
student_copy=student
print("Copies dict to another Dict  using = operator ")
print(student)
print(student_copy)

# value is modified 
student["name"]="Krish2"
print("Updated some value in main dict  ")
print(student)
print(student_copy)

print("==============================================================")

# Note : when we make the copy it should not change the value it should have the original value and 
# just make the copy of it another variable and memory is allocated here 
student_copy1=student.copy() ## shallow copy
print("Printing Dict using shallow copy (copy()) ")
print(student_copy1)
print(student)

student["name"]="Ram"
print("Updated the values of one key")
print(student_copy1)
print(student)

```

```python
### Iterating Over Dictionaries
## You can use loops to iterate over dictionatries, keys,values,or items
student={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}
## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for value in student.values():
    print(value)
```

```python
## Iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries
students={
    "student1":{"name":"Krish","age":32},
    "student2":{"name":"Peter","age":35}
}
print(students)
```

```python
## Access nested dictionaries elementss
print(students["student2"]["name"])
print(students["student2"]["age"])
print(students["student1"]["name"])
```

```python
students.items()
```

```python
## Iterating over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comphrehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Condition dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples

## USe a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for i in numbers:
    if i in frequency:
        frequency[i]+=1
    else:
        frequency[i]=1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```

```python
# count the frequency of each character

s="india is my country"
freq={}

for ch in s:
    freq[ch]=freq.get(ch,0)+1

print(freq)
```
