I. intersection()

 1. Returns a new set with common elements
 2. Does NOT modify original sets


II. intersection_update()

 1. Updates the original set itself
 2. Keeps only common elements


Imp : When to Use What?
1. Use intersection() -- when you want to keep original sets safe
2. Use intersection_update() --  when you want to directly modify the set for efficiency

```python
s1={10,20,30,40,50}
s2={40,50,60,70}

#intersection will only give you common element
print(s1.intersection(s2))
print(s1)
print(s2)

# intersection update will find common element and add that element in s1 
s1.intersection_update(s2)
print(s1)
```

I. difference()

Returns elements that are in set A but NOT in set B

Meaning:
Remove common elements
Keep only elements unique to first set


II. symmetric_difference()

Returns elements that are in either set, but NOT in both

Meaning:
Remove common elements
Keep elements from both sets except common ones

Quick Understanding
1. difference() 
“What is in A but not in B?”
2. symmetric_difference() 
“What is different between A and B?”

```python
s1={10,20,30}
s2={20,30,40,50}

#difference
print(s2.difference(s1))
print(s1.difference(s2))
```

```python
a={1,2,3}
b={2,3,4}

#symmetric difference
print(a.symmetric_difference(b))
```

```python
# find pair with given sum

arr=[1,2,3,4,5]
target=6

seen=set()

for num in arr:
    if target-num in seen:
        print(num,target-num)
    seen.add(num)
```

```python
# find duplicate

lst=[1,2,3,2,3,4,5,4,3]

seen=set()

for i in lst:
    if i in seen:
        print("Duplicate Element : ",i)
    else:
        seen.add(i)

print(seen)
```
