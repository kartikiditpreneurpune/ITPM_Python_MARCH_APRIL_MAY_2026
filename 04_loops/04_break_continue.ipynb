```python
# skip number 5 

for i in range(1,11):
    if i==5:
        continue
    print(i)
```

```python
# skip the even numbers
for i in range(1,11):
    if i%2==0:
        continue
    print(i)
```

```python
# skip negative number

nums=[10,-4,5,-5,8,-9]

for i in nums:
    if i<0:
        continue
    print(i)
```

```python
# skip the failed stduents

marks=[20,40,50,80,14,15]

for i in marks:
    if i<35:
        continue
    print("Pass Student MArks :: ",i)
```

Pass Statement :
1. continue and pass are different
2. Continue will skip current iteration and will continue with next iteration
3. pass is used when a statement is required syntactically , but you don't have the code you want to execute till now
