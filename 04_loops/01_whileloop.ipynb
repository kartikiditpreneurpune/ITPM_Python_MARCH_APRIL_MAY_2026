```python
# sum of numbers
numrange=int(input("Enter Range of number to ADD :: "))

i=1
total=0

while i<=numrange:
    total+=i
    i+=1

print("Sum is :: ",total)
```

```python
# ATM Correct Pin Verification -- 3 times

correct_pin=1234
attempt=1

while attempt<=3:
    pin=int(input("Enter ATM Pin :: "))

    if pin==correct_pin:
        print("Login to ATM is Successful !! ")
        break

    print("Wrong PIN ")
    attempt+=1

if attempt>3:
    print("Account Blocked !!! ")
```

```python
# Shopping Cart 
cart=[]

item=""

while item!="done":
    item=input("Enter Item Name (or done for finishing the shopping ):: ")

    if item!="done":
        cart.append(item)


print("Items is Cart is :: ",cart)
```
