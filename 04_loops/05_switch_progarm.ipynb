```python
# atm Menu

choice=int(input("Enter Your Choice :: 1. Balance Enquiry 2. Deposit 3. Withdraw"))

match choice:
    case 1:
        print("Balance check ")
    case 2:
        print("Deposit Money")
    case 3:
        print("Withdraw Money")
    case _:
        print("Invalid CHoice ")
```

```python
# check the traffic signal options

signal=input("Enter Colour of Signal [red,green,yellow] :: ")

match signal:
    case "red":
        print("STOP")
    case "green":
        print("GOOOOO")
    case "yellow":
        print("Wait")
    case _:
        print("Invalid Signal")
```
