```python
# sqaure patter of 5 *5

for i in range(5):
    for j in range(5):
        print("*",end=" ")
    print()
```

```python
# Right Triangle using *

for i in range(1,6):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
# Right Triangle using number

for i in range(1,6):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
# Reverse Star pattern 

for i in range(5,0,-1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
# Reverse Number pattern 

for i in range(5,0,-1):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

```python
# same number trinagle pattern

for i in range(1,6):
    for j in range(i):
        print(i,end=" ")

    print()
```

```python
# alphabet Triangle pattern

for i in range(1,6):
    for j in range(i):
        print(chr(64+i),end=" ")
    print()
```

```python
# Continous Number Printtng

num=1

for i in range(1,6):
    for j in range(i):
        print(num,end=" ")
        num=num+1
    print()
```

```python
# student marks report 

students=["Rahul","Raj","Ram"]

for i in students:
    print("Student name is :: ",i)

    for subject in ["Java","Python","MERN"]:
        print("Subject is :: ",subject)

    print()
```

```python
# Print a Cinema Theatre Seating Arrangement

for row in range(1,4):
    print("Row number is ::",row)

    for seat in range(1,6):
        print("  Seat :: ",seat,end=" ")

    print("\n ")
```
