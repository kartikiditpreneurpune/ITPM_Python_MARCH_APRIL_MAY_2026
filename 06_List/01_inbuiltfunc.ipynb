```python
#Nested List 

lst=[[1,2,3],[4,5,6],[1,"ram",90.90]]

print(lst)

print(lst[0])

print(lst[0][2])
print(lst[2][1])

print(lst[0][0:3])

print(lst[2][1:3])
```

```python
# append : is used to add elements from last index

lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List :: ",lst)

# append 
element=int(input("Enter Element to add :: "))
lst.append(element)

print("Updated List :: ",lst)
```

```python
# extend == will allow you to add multiple elements in list 
# extends == allow to add one list to other


lst1=[]
lst2=[]

n=int(input("Enter Count of Elements  for List 1 :: "))
for i in range(n):
    num=int(input("Enter Elements :: "))
    lst1.append(num)

n=int(input("Enter Count of Elements  for List 2 :: "))
for i in range(n):
    num=int(input("Enter Elements :: "))
    lst2.append(num)

print("List 1 :: ",lst1)
print("List 2 :: ",lst2)

lst1.extend(lst2)
print("Updated List 1 :: ",lst1)

```

```python
# insert == used to add element at specific index/ position

lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print(lst)

ind=int(input("Enter Index Number :: "))
val=int(input("Enter Value to add :: "))

lst.insert(ind,val)

print("Updated List is :: ",lst)
```

```python
# delete the value : remove()
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print(lst)

val=int(input("enter Value to delete :: "))

if val in lst:
    lst.remove(val)
else:
    print("Value Not FOund ")

print("Updated List :: ",lst)
```

```python
# delete the element using specific index  : pop()
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print(lst)

ind=int(input("enter index to delete element :: "))

if ind<len(lst):
    lst.pop(ind)
else:
    print("Index Not FOund ")

print("Updated List :: ",lst)
```

```python
# delete the element using   : pop()

lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print(lst)
res=lst.pop()
print("Updated List :: ",lst)
print("Removed element :: ",res)
```

```python
# find index of element
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print(lst)

val=int(input("Enter Value to find Index :: "))

if val in lst:
    print("ELement find at index :: ",lst.index(val))
else:
    print("Value not Found")

```

```python
# find the occurance of elements

lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print(lst)

val=int(input("enter element to count the frequeny :: "))
print("Occurance of elements is :: ",lst.count(val))
```

```python
# sort the list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print(lst)

# ascending 
lst.sort()
print("Sorted list in ascending order is  :: ",lst)

#descending 
lst.sort(reverse=True)
print("Sorted List in descending order is :: ",lst)
```

```python
# reverse the list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print(lst)

lst.reverse()
print("After Reversed list is :: ",lst)
```

```python
# copy the list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List is ::",lst)

new_list=lst.copy()

print("Copied List is :: ",new_list)
```

```python
# calculate length
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List is ::",lst)

print("LEngth of List is :: ",len(lst))
```

```python
# min,max,sum
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List is ::",lst)

print("Minimum element in list is :: ",min(lst))
print("Maximum element in list is :: ",max(lst))
print("Sum of all element in list is :: ",sum(lst))
```
