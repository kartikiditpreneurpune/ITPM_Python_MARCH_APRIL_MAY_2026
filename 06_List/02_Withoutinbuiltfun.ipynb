```python
# find the max element from list
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List is ::",lst)

largest=lst[0]

for i in lst:
    if i>largest:
        largest=i

print("Largest element in list is :: ",largest)
```

```python
# remove the dupliaction from list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List is ::",lst)

no_dupliactes=[]

for i in lst:
    if i not in no_dupliactes:
        no_dupliactes.append(i)


print("List without Duplicatiosn is :: ",no_dupliactes)
```

```python
# reverse a list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List is ::",lst)

rev=[]
for i in range(len(lst)-1,-1,-1):
    rev.append(lst[i])


print("Reversed List is :: ",rev)
```

```python
lst=[10,20,30,40,50]
i=len(lst)-1
rev=[]

while i>=0:
    rev.append(lst[i])
    i-=1

print("Original List: ",lst)
print("Reverses List :: ",rev)
```

```python
# separte even / odd from list
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List is ::",lst)

even=[]
odd=[]

for i in lst:
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)

print("Even List is :: ",even)
print("Odd List is :: ",odd)
```

```python
# even and oddd index
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

print("Original List is ::",lst)

even_index=[]
odd_index=[]

for i in range(len(lst)):
    if i%2==0:
        even_index.append(lst[i])
    else:
        odd_index.append(lst[i])

print("Even Index Elemenets :: ",even_index)
print("Odd Index eLements :: ",odd_index)
```

```python
# square of elements in list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

#singlr line logic
square=[num*num for num in lst]

print("Original list : ",lst)
print("Square of element list :: ",square)    
```

```python
# cube of elements in list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

#singlr line logic
cube=[num**3 for num in lst]

print("Original list : ",lst)
print("Cube of element list :: ",cube)    
```

```python
# even elements in list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

#singlr line logic
even=[num for num in lst if num%2==0]

print("Original list : ",lst)
print("Even element list :: ",even) 
```

```python
# odd elements in list 
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

#singlr line logic
odd=[num for num in lst if num%2!=0]

print("Original list : ",lst)
print("Even element list :: ",odd) 
```

```python
# matrix -- 2D array 

matrix=[]
rows=int(input("Enter Number of Rows :: "))
cols=int(input("Enter Number of Columns :: "))

for i in range(rows):
    row=[]
    for j in range(cols):
        val=int(input("Enter Value :: "))
        row.append(val)
    matrix.append(row)

print("Matrix is ::" )

for r in matrix:
    for c in r:
        print(c,end=" ")
    print()
```

```python
# fetching duplicates
lst=[]

n=int(input("Enter Count of Elements :: "))

for i in range(n):
    num=int(input("Enter Elements :: "))
    lst.append(num)

dupliactes=[]
for i in lst:
    if lst.count(i)>1 and i not in dupliactes:
        dupliactes.append(i)

print("Dupliactes elements are :: ",dupliactes)
```
