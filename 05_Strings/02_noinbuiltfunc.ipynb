```python
# reverse the string 
str=input("Enter String :: ")

rev=""

for ch in str:
    rev=ch+rev

print("Reversed String is :: ",rev)
```

```python
# find the max occurance of characers

str=input("Enter String :: ")
max_char=""
max_count=0

# traversing the string 
for ch in str:
    count=0
    # have the max count
    for c in str:
        if ch==c:
            count+=1

    if count>max_count:
        max_count=count
        max_char=ch 
    
print("MAximum Occuring Character is :: ",max_char)
print("Maximum Count is :: ",max_count)
```

```python
# count the occurance of character
str=input("Enter the string :: ")
ch=input("Enter Character to check Occurance:: ")
count=0

for c in str.lower():
    if c==ch:
        count+=1

print("Occurance of Character is :: ",count)
```

```python
# remove the spaces from string 
str=input("Enter the string :: ")
result=""

for ch in str:
    if ch!=" ":
        result+=ch

print("String without space is :: ",result)
```

```python
# copy one string to another 
str=input("Enter the string :: ")
copy_str=""

for ch in str:
    copy_str+=ch

print("Original String is :: ",str)
print("Copied String is :: ",copy_str)
```

```python
# Remove Duplications from String 
str=input("Enter the string :: ")
result=""

for ch in str:
    if ch not in result:
        result+=ch

print("String without duplication :: ",result)
```

```python
# compare two strings without == opeartor 

s1=input("Enter String 1 :: ")
s2=input("Enter String 2 :: ")

if len(s1)!=len(s2):
    print("String are not equal ")
else:
    equal=True
    for i in range(len(s1)):
        if s1[i]!=s2[i]:
            equal=False
            break
    if equal:
        print("Strings are Equal ")
    else:
        print("Strings are not Equal ")
```
