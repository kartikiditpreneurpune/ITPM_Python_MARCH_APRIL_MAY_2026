```python
# check the balance and accordingly need to withdraw the money where limit is 2k 

balance=float(input("enter Balance Amount :: "))
amount=float(input("Enter AMount to withdraw :: [ limit is 2K ]"))

print("Balance is :: ",balance)

if amount<=balance and amount<=2000:
    balance=balance-amount
    print("Withdrawn Amount Successfully !!! ")
    print("Balance is :: ",balance)
else:
    print("Failed To Withdraw !!! Limit is 2000/-")
```

```python
# Consider done shopping if worth Rs. 5000/- or above , then cashback of 500 is applicable 

bill_amount=float(input("Enter Bill AMount :: "))

if bill_amount>=5000:
    bill_amount=bill_amount-500
    print("CashBack is Applicable !!! ")
    print("Final Bill is :: Rs. ",bill_amount)
else:
    print("No Cashback Is Applicable ")
    print("Final Bill is :: Rs. ",bill_amount)
```
