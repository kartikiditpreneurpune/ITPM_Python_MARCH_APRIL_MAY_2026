```python
# create a grade system

marks=int(input("Enter Marks :: "))
if marks>=90 and marks<=100:
    print("Grade A")
elif marks>=75 and marks<90:
    print("Grade B")
elif marks>=35 and marks<75:
    print("Grade C")
else:
    print("Fail")
```

```python
# accept the colour of signal 

signal=input("Enter Signal Colour :: ")
if signal=="red":
    print("Stop")
elif signal=="yellow":
    print("Slow Down The Vehicle !!! ")
elif signal=="green":
    print("GO ")
else:
    print("Please give valid colour signal - red, green,yellow")
```

```python
# accept the bill and then accordingly i need to give the disocunt

bill=float(input("Enter Bill Amount :: "))

if bill>=10000:
    discount=bill*0.25
    bill=bill-discount
    print("Discount is :: ",discount)
    print("Final Bill is :: ",bill)
elif bill>=5000 and bill<10000:
    discount=bill*0.15
    bill=bill-discount
    print("Discount is :: ",discount)
    print("Final Bill is :: ",bill)
else:
    print("NO Discount is applicable")
    print("Final Bill is :: ",bill)
```

```python
# accept the salary and decide the tax slab

salary=float(input("Enter Salary :: "))

if salary<=30000:
    print("No Tax")
elif salary<=70000:
    print("5 % tax")
else:
    print("10 % Tax ")
```
